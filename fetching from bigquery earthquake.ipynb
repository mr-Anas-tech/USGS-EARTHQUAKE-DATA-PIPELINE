```python
df['sensor_reliability_flag'] = np.where(
    (df['total_stations_used'] < 12) & (df['azimuthal_gap'] > 180), 
    'Low Reliability (Risky)', 
    'High Reliability (Solid)'
)
```

```python
network_performance = df.groupby('sensor_reliability_flag').agg(
    total_events=('earthquake_id', 'count'),
    avg_pipeline_delay_mins=('pipeline_reporting_delay_minutes', 'mean'),
    avg_rms_travel_time=('rms_travel_time', 'mean')
).reset_index()
network_performance
```

```python
depth_severity_matrix = pd.crosstab(
    df['depth_category'], 
    df['magnitude_severity_class'], 
    margins=True
).sort_values(by='All', ascending=False)
depth_severity_matrix 
```

```python
tsunami_impact = df.groupby('has_tsunami_risk').agg(
    avg_significance=('significance_score', 'mean'),
    max_reported_intensity=('max_reported_intensity', 'max'),
    avg_felt_reports=('total_felt_reports', 'mean')
).reset_index()
tsunami_impact
```

```python
hourly_trends = df.groupby('event_hour').agg(
    event_count=('earthquake_id', 'count'),
    avg_magnitude=('magnitude', 'mean'),
    total_public_reports=('total_felt_reports', 'sum')
).reset_index().sort_values(by='event_count', ascending=False)
hourly_trends
```
