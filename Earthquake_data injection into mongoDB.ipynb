```python
import requests
from pymongo import MongoClient
import  time
import certifi
client=MongoClient("mongodb://localhost:27017/")
db=client["earthquick_db"]
collection=db["raw_earthquick_data"]
# api
url="https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson"
print("earthquick  data is streaming")
while True:
    try:
        response=requests.get(url)
        data=response.json()
        events=  data.get("features",[])
        new_record_count=0
        for event in events:
            unique_id= event.get("id")
            if not collection.find_one({"id": unique_id}):
                collection.insert_one(event)
                new_record_count+=1
                if new_record_count> 0:
                   print(f"{new_record_count} loaded into mongo")
                else:
                   print(f"checking_api")
                time.sleep(30)
    except Exception as  e:
        print(f"error occurred:{e}")
        time.sleep(10)
            
            
```
